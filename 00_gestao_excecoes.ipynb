```python
import pandas as pd
import os
from google.colab import drive

# 1. Montar o Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. Definir o caminho na raiz do seu DATA_LAKE
# Assim fica fácil de qualquer notebook (Bancos, FIIs, etc) achar
path_excecoes = '/content/drive/MyDrive/DATA_LAKE/gestao_excecoes.csv'

# 3. Criar ou Carregar a estrutura
if os.path.exists(path_excecoes):
    df_excecoes = pd.read_csv(path_excecoes)
    print("📖 Planilha existente carregada. Adicionando nova exceção...")
else:
    df_excecoes = pd.DataFrame(columns=['Ativo', 'Data', 'Tipo', 'Valor_Correto'])
    print("🆕 Criando nova planilha de exceções.")

# 4. Adicionar o ajuste do XPML11 (Tratando o erro de split/distribuição)
# Exemplo: Se o dado veio 202% (2.02), o valor correto seria algo como 0.5% (0.005)
novo_ajuste = {
    'Ativo': 'XPML11',
    'Data': '2026-01-31',
    'Tipo': 'ERRO_SPLIT',
    'Valor_Correto': 0.005
}

# Evitar duplicados: só adiciona se essa combinação Ativo+Data não existir
if not ((df_excecoes['Ativo'] == novo_ajuste['Ativo']) & (df_excecoes['Data'] == novo_ajuste['Data'])).any():
    df_excecoes = pd.concat([df_excecoes, pd.DataFrame([novo_ajuste])], ignore_index=True)
    # 5. Salvar
    df_excecoes.to_csv(path_excecoes, index=False)
    print(f"✅ Exceção registrada para {novo_ajuste['Ativo']} em {novo_ajuste['Data']}")
else:
    print(f"⚠️ Essa exceção já estava registrada.")

print(f"📁 Localização: {path_excecoes}")
display(df_excecoes)
```
