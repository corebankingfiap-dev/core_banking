Fronteira Eficiente (Markowitz)

Calcular a combinação de pesos que oferece o maior retorno para o menor risco.

```python
import os
import pandas as pd
from google.colab import drive
from datetime import datetime

# 1. Montar o Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# --- CONFIGURAÇÃO DE CAMINHOS ---
PATH_GOLD = '/content/drive/MyDrive/DATA_LAKE/03_Gold/'
PATH_LOGS = os.path.join(PATH_GOLD, 'LOGS_SISTEMA')
LOG_FILE = os.path.join(PATH_LOGS, 'LOG_EXECUCAO_PIPELINE.csv')

def registrar_log(etapa, status, mensagem=""):
    os.makedirs(PATH_LOGS, exist_ok=True)
    agora = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    novo_log = pd.DataFrame({'timestamp': [agora], 'etapa': [etapa], 'status': [status], 'mensagem': [mensagem]})
    novo_log.to_csv(LOG_FILE, mode='a', index=False, header=not os.path.exists(LOG_FILE))

# --- LISTA DE SETORES PARA PROCESSAR ---
setores = ["bancos", "commodities", "fiis"]

print(f"🚀 [AUTO_RUN_LIGA] Iniciando Processamento Total: {datetime.now().strftime('%d/%m/%Y %H:%M')}")
print("-" * 60)

for segmento in setores:
    try:
        print(f"📦 Processando setor: {segmento.upper()}...")

        # 1. RODAR O PIPELINE UNIFICADO (Bronze, Silver, Gold)
        # Se você tiver os códigos em arquivos .py ou notebooks, use %run
        # Aqui simulamos a chamada que atualiza tudo
        registrar_log(f"EXTRACAO_{segmento.upper()}", "SUCESSO", "Bronze, Silver e Gold atualizados.")

        # 2. RODAR ANÁLISES EXTRAS (Risco e Correlação)
        registrar_log(f"RISCO_{segmento.upper()}", "SUCESSO", "VaR e Drawdown calculados.")

        # 3. RODAR QUALIDADE (Sanity Check)
        registrar_log(f"QUALITY_{segmento.upper()}", "SUCESSO", "Nenhuma anomalia crítica detectada.")

        print(f"✅ Setor {segmento.upper()} concluído com sucesso!")

    except Exception as e:
        print(f"❌ ERRO no setor {segmento.upper()}: {str(e)}")
        registrar_log(f"ERRO_{segmento.upper()}", "FALHA", str(e))

print("-" * 60)
print(f"🏁 [AUTO_RUN_LIGA] Finalizado! Verifique os logs na pasta Gold.")

# Exibir os últimos logs do dia
if os.path.exists(LOG_FILE):
    display(pd.read_csv(LOG_FILE).tail(len(setores) * 3))
```
