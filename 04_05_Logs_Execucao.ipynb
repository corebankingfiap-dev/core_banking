Cria um arquivo de log centralizado na sua camada Gold que registre toda vez que o pipeline rodar, marcando o sucesso, o erro e o tempo de execução

```python
import pandas as pd
import os
from datetime import datetime
from google.colab import drive

# 1. Montar o Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. CONFIGURAÇÃO DE CAMINHO (Padronizado com seu Data Lake)
PATH_GOLD = '/content/drive/MyDrive/DATA_LAKE/03_Gold/'
PATH_LOGS = os.path.join(PATH_GOLD, 'LOGS_SISTEMA') # Criando uma subpasta para não poluir a Gold
LOG_FILE = os.path.join(PATH_LOGS, 'LOG_EXECUCAO_PIPELINE.csv')

def registrar_log(etapa, status, mensagem=""):
    """
    Registra o status de execução de uma etapa do pipeline de forma incremental.
    """
    os.makedirs(PATH_LOGS, exist_ok=True)
    agora = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    novo_log = {
        'timestamp': [agora],
        'etapa': [etapa],
        'status': [status],
        'mensagem': [mensagem]
    }

    df_novo = pd.DataFrame(novo_log)

    # Uso de 'header=not os.path.exists' para economizar memória e evitar ler o CSV inteiro só para anexar
    # Modo 'a' (append) é muito mais rápido para arquivos de log longos
    df_novo.to_csv(LOG_FILE, mode='a', index=False, header=not os.path.exists(LOG_FILE))
    print(f"📝 Log registrado: {etapa} -> {status}")

# --- EXEMPLO DE USO INTEGRADO ---

try:
    print("🚀 Iniciando atualização geral do Data Lake...")

    # Aqui você colocaria a chamada dos seus outros códigos
    # Exemplo: %run 01_Bronze_Bancos.ipynb

    registrar_log("PIPELINE_GERAL", "SUCESSO", "Bancos, Commodities e FIIs atualizados.")

except Exception as e:
    registrar_log("PIPELINE_GERAL", "FALHA", f"Erro crítico: {str(e)}")

# Exibir os últimos logs para conferência rápida
if os.path.exists(LOG_FILE):
    print("\n--- HISTÓRICO RECENTE DE EXECUÇÃO ---")
    display(pd.read_csv(LOG_FILE).tail(10))
```
