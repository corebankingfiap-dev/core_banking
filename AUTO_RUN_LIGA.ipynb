Executa comandos de forma automatizada

```python
from google.colab import drive
import os
from datetime import datetime
import pandas as pd

# 1. Conexão com o Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. Configurações de Caminho (Ajustado para o seu padrão de pastas)
PATH_NOTEBOOKS = "/content/drive/MyDrive/Colab Notebooks/"
PATH_LOGS = "/content/drive/MyDrive/DATA_LAKE/03_Gold/LOGS_SISTEMA/"
LOG_FILE = os.path.join(PATH_LOGS, 'LOG_EXECUCAO_PIPELINE.csv')

def registrar_log(etapa, status, mensagem=""):
    os.makedirs(PATH_LOGS, exist_ok=True)
    agora = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    novo_log = pd.DataFrame({'timestamp': [agora], 'etapa': [etapa], 'status': [status], 'mensagem': [mensagem]})
    novo_log.to_csv(LOG_FILE, mode='a', index=False, header=not os.path.exists(LOG_FILE))

def executar_fluxo(arquivo):
    caminho_total = os.path.join(PATH_NOTEBOOKS, arquivo)
    if os.path.exists(caminho_total):
        print(f"🚀 Rodando: {arquivo}")
        try:
            # O comando %run executa o notebook como se fosse um script
            %run "{caminho_total}"
            registrar_log(arquivo, "SUCESSO", "Notebook executado sem erros.")
            print(f"✅ {arquivo} OK")
        except Exception as e:
            registrar_log(arquivo, "FALHA", str(e))
            print(f"❌ Erro em {arquivo}: {e}")
    else:
        registrar_log(arquivo, "NAO_ENCONTRADO", "Caminho especificado não existe.")
        print(f"⚠️ Arquivo não encontrado: {arquivo}")

# 3. LISTA MESTRE - Ordem de execução (Bronze -> Silver -> Gold -> Reports)
fluxo_da_liga = [
    "01_Data_Lake_Bancos.ipynb",
    "01_Data_Lake_Commodities.ipynb",
    "01_Data_Lake_FIIs.ipynb",
    "03_Data_Lake_Analise_Quant_Bancos.ipynb",
    "03_Data_Lake_Analise_Quant_Commodities.ipynb",
    "03_Data_Lake_Analise_Quant_FIIs.ipynb",
    "00_Data_Lake_Consolidação_Portfolio.ipynb"
]

print("🏁 INICIANDO ATUALIZAÇÃO TOTAL DA LIGA FINANCEIRA\n" + "="*50)

for nb in fluxo_da_liga:
    executar_fluxo(nb)

print("\n" + "="*50 + "\n🏆 PIPELINE FINALIZADO! Verifique os LOGS na camada GOLD.")
```
